```python


import os
import random
import time

# ---------- Helper functions ----------
def save_name(name):
    with open("user_data.txt", "w") as f:
        f.write(name)

def load_name():
    if os.path.exists("user_data.txt"):
        with open("user_data.txt", "r") as f:
            return f.read().strip()
    return None

def detect_mood(text):
    """Simple keyword-based mood detector (positive / negative / neutral)."""
    t = text.lower()
    positive_words = ["happy","good","great","awesome","fine","excited","joy","glad","proud","better"]
    negative_words = ["sad","alone","bad","tired","upset","angry","low","depressed","worried","stressed","anxious","hate","hurt","lonely"]
    # explicit check for strong negative/self-harm language
    crisis_words = ["kill myself","suicide","end my life","hurt myself","want to die","die by suicide","i'll die"]
    if any(kw in t for kw in crisis_words):
        return "crisis"
    if any(kw in t for kw in positive_words):
        return "positive"
    if any(kw in t for kw in negative_words):
        return "negative"
    return "neutral"

def gentle_sleep_nudge(name):
    opts = [
        f"Maybe go nap a little, {name}. Let your chaotic world pause for a minute ðŸ˜´",
        f"Sleep helps more than a million pep talks, {name}. Give yourself that tiny break ðŸŒ™",
        f"Put the world on snooze for an hour and pretend youâ€™ve got all the time in the morning ðŸ’«"
    ]
    return random.choice(opts)

def gentle_action_nudge(name):
    opts = [
        f"Write one small thing that went okay today â€” even tiny wins count, {name} ðŸŒ¤ï¸",
        f"Try a deep breath with me: in for 4, hold 2, out for 6. Repeat twice, {name} ðŸŒ¿",
        f"Drink a glass of water, stretch a bit â€” small moves help steady the mind ðŸ’§"
    ]
    return random.choice(opts)

def crisis_response():
    lines = [
        "Iâ€™m really sorry youâ€™re feeling so overwhelmed. Iâ€™m not a human therapist, but I care and want you to be safe.",
        "If youâ€™re thinking about hurting yourself, please consider contacting a nearby emergency number or a crisis line right now.",
        "If you are in India you can call/WhatsApp iCALL (9152987821) or look up your local emergency services. If elsewhere, please contact your local crisis services or emergency number.",
        "You deserve immediate human help â€” please reach out to someone you trust or a professional right now."
    ]
    return "\n".join(lines)

# ---------- Welcome & name memory ----------
print("ðŸŒ¸ Welcome to Repose â€” your comfort zone ðŸŒ¸")
print("A little peace corner in your chaotic world.")
time.sleep(0.4)

stored_name = load_name()
if stored_name:
    user_name = stored_name
    print(f"Repose: Welcome back, {user_name} â€” glad youâ€™re here ðŸŒ¼")
else:
    # playful prompt
    user_name = input("Repose: Hey, what's your name, you intelligent creature? ðŸ‘€\nYou: ").strip().capitalize()
    if user_name == "":
        user_name = "friend"
    save_name(user_name)
    print(f"Repose: Nice to meet you, {user_name}! So good to have you here ðŸ’«")

time.sleep(0.2)
print("\n(Repose is listening. Type 'bye' / 'exit' / 'quit' to end the chat.)\n")

# ---------- Session state ----------
mood_counts = {"positive":0, "neutral":0, "negative":0}
turn_count = 0

# ---------- Conversation loop ----------
while True:
    you = input("You: ").strip()
    if not you:
        # skip empty input politely
        print(f"Repose: It's okay to take your time, {user_name}. When you're ready, I'm here.")
        continue

    lower = you.lower()
    # exit
    if lower in ["bye","exit","quit"]:
        print(f"Repose: Take care, {user_name}. You did well today â€” small steps matter ðŸŒ™")
        break

    # crisis check (self-harm language)
    mood_check = detect_mood(you)
    if mood_check == "crisis":
        print("\nRepose:", crisis_response())
        # After crisis response, encourage contacting human help and exit politely
        print("\nRepose: I'm staying with you in this chat as long as you need, but please consider reaching out to local emergency or a trusted person right now.")
        # Do not continue the normal loop: let user decide, but we continue the loop so they can type again
        continue

    # special greeting handling
    if lower in ["hi","hello","hey"]:
        greetings = [
            f"Hi {user_name}! How are you doing today? Still rocking your world as always? âœ¨",
            f"Hey {user_name}! Look who showed up â€” my favorite human ðŸŒ¸",
            f"Hello sunshine â˜€ï¸, howâ€™s your universe treating you?",
            f"Heeey {user_name}! Ready to sprinkle a little calm and charm? ðŸŒ¿",
            f"Yo {user_name}! You bring the vibe every time you appear ðŸ˜Œ"
        ]
        print("Repose:", random.choice(greetings))
        continue

    # detect mood normally
    mood = detect_mood(you)
    if mood not in ["positive","neutral","negative"]:
        mood = "neutral"
    mood_counts[mood] += 1
    turn_count += 1

    # positive responses
    if mood == "positive":
        responses = [
            f"Thatâ€™s lovely to hear, {user_name}! You deserve this brightness ðŸŒ¼",
            f"You're glowing today, {user_name}. Keep that warmth with you âœ¨",
            f"Yay! I'm so happy you're feeling good â€” celebrate the small joys ðŸ¥³"
        ]
        print("Repose:", random.choice(responses))

    # negative responses (psychology-based + gentle humour)
    elif mood == "negative":
        # core comforting messages (acknowledge -> normalize -> reframe)
        comfort_msgs = [
            f"I can tell today feels heavy, {user_name}. Thatâ€™s okay â€” feeling is allowed ðŸ’›",
            f"You're allowed to rest. This doesnâ€™t make you weak, {user_name} â€” it makes you human ðŸŒ¿",
            f"Feelings come and go like weather. This cloud will pass too, {user_name} â˜ï¸",
            f"You've handled hard things before, {user_name}. Weâ€™ll take this one small step at a time ðŸ’«"
        ]
        print("Repose:", random.choice(comfort_msgs))

        # add a gentle humorous nudge sometimes
        if random.random() > 0.55:
            print("Repose:", gentle_sleep_nudge(user_name))
        elif random.random() > 0.6:
            print("Repose:", gentle_action_nudge(user_name))

    # neutral responses
    else:  # neutral
        neutral_msgs = [
            f"Just taking things as they come, huh {user_name}? That's okay ðŸŒ¸",
            f"Some days are quiet, and that can be a kind of healing â€” tell me if you want to talk more â˜ï¸",
            f"Want a tiny exercise? Breathe in for 4, hold 2, out for 6. I'm with you ðŸŒ¿"
        ]
        print("Repose:", random.choice(neutral_msgs))

    # small chance to offer a calm micro-activity or suggestion
    if random.random() > 0.8:
        micro = random.choice([
            "Repose: Want a tiny breathing exercise? (type 'breathe' to do it now)",
            "Repose: If you like, I can share a short calming quote â€” say 'quote'.",
            "Repose: If you're up for it, tell me one small win from today."
        ])
        print(micro)

    # handle special micro commands
    if lower == "breathe":
        print("Repose: Let's breathe together â€” in... 2... 3... 4... hold... out... 1... 2... done. How do you feel?")
    if lower == "quote":
        print("Repose: \"This too shall pass â€” and you're stronger than you think.\" ðŸŒ¼")

# ---------- Session summary ----------
print("\n--- Session Summary ---")
print(f"Turns this session: {turn_count}")
print(f"Positive messages: {mood_counts['positive']}")
print(f"Neutral messages: {mood_counts['neutral']}")
print(f"Negative messages: {mood_counts['negative']}")
print("-----------------------\n")

print("Repose: If you want, next time I can remember how you were feeling and check in. Take gentle care, see you soon âœ¨")
```
